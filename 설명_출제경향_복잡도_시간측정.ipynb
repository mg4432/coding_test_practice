## 복잡도
 - 시간 복잡도 : 특정한 크기의 입력에 대하여 알고리즘의 수행 시간 분석
 - 공간 복잡도 : 특정한 크기의 입력에 대하여 알고리즘의 메모리 사용량 분석

동일한 기능을 수행하는 알고리즘이 있다면 일반적으로 복잡도가 낮을수록 좋은 알고리즘


# 복잡도

## 시간 복잡도 예시1

```python
array = [3, 5, 1, 2, 4] # 리스트 객체 할당
summary = 0 # 합계 저장할 변수

# 모든 데이터를 하나씩 확인하면서 합계 계산
for x  in array : 
    summary += x
    
print(summary)
```

수행 시간은 데이터의 개수 N에 비례 **$O(N)$**

 - 모든 데이터를 하나씩 확인하며 계산하기 때문 

## 시간 복잡도 예시2 - 2중 반복문 이용

```python
array = [3, 5, 1, 2, 4]

for i in array :
    for j in array : 
        temp = i*j
        print(temp)
```

수행 시간은 데이터의 개수 N에 비례 **$O(N^2)$**

모든 2중 반복문의 시간복잡도가 $O(N^2)$인 것은 아님
- 소스코드가 내부적으로 다른 함수를 호출한다면 그 함수의 시간복잡도까지 고려해야 한다

일반적으로 CPU 기반의 개인 컴퓨터나 채점용 컴퓨터에서 연산 횟수가 5억을 넘어가는 경우 :
 - Python의 경우 통상 5\~15초의 시간이 소요(C언어는 통상 1\~3초)
 - PyPy의 경우 C언어보다 빠르게 동작하기도 한다

$O(N^3)$의 알고리즘을 설계한 경우 N의 값이 5000을 넘으면 얼마나 걸릴 것인가?

```python
5000**3 # 1250억
```

파이썬이 1초에 약 5,000만 번의 계산을 처리할 수 있다고 하면 2,500초 정도 소요된다고 생각할 수 있음

채점 서버에서는 파이썬이 1초에 약 2,000만 번의 계산만을 처리할 수 있다고 생각하고 계산하는 것을 추천

## 요구사항에 따라 적절한 알고리즘 설계
- 가장 먼저 확인해야 할 내용은 시간제한(수행시간 요구사항)
- 시간제한이 1초인 문제를 만났을 때, 일반적인 기준 
 - N의 범위가 500인 경우 : 시간 복잡도가 $O(N^3)$인 알고리즘을 설계하면 문제를 풀 수 있다.
 - N의 범위가 2,000인 경우 : 시간 복잡도가 $O(N^2)$인 알고리즘을 설계하면 문제를 풀 수 있다.
 - N의 범위가 100,000인 경우 : 시간 복잡도가 $O(NlogN)$인 알고리즘을 설계하면 문제를 풀 수 있다.
 - N의 범위가 10,000,000인 경우 : 시간 복잡도가 $O(N)$인 알고리즘을 설계하면 문제를 풀 수 있다.

## 수행시간 측정 소스코드 예제

```python
import time 
start_time = time.time() # 측정 시작

# 프로그램 소스 코드 

end_time = time.time() # 측정 종료
print("time :", end_time - start_time) # 수행 시간 출력
```

```python
import time 
start_time = time.time() # 측정 시작

array = [3, 5, 1, 2, 4]

for i in array :
    for j in array : 
        temp = i*j
        print(temp)
        
end_time = time.time() # 측정 종료
print("time :", end_time - start_time) # 수행 시간 출력
```
