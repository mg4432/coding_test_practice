```python
n = int(input()) 

count = 1 
stack = [] 
result = [] 

for i in range(1, n+1) :
    data = int(input())
    while count <= data :
        stack.append(count)
        count += 1 
        result.append('+')
        
    if stack[-1] == data : 
        stack.pop()
        result.append('-')
        
    else : 
        print('NO')
        break
        
print('\n'.join(result))
        
```

```python
test_case = int(input())

for _ in range(test_case) : 
    n, m = list(map(int, input().split(' ')))
    queue = list(map(int, input().split(' ')))
    queue = [(i, idx) for idx, i in enumerate(queue)] # 인덱스 추가
    
    count = 0 
    while True : 
        if queue[0][0] == max(queue, key = lambda x : x[0])[0] :
            count += 1 
            if queue[0][1] == m : 
                print(count)
                break
            else :
                queue.pop(0)
        else :
            queue.append(queue.pop(0))
```

```python
test_case = int(input()) 

for _ in range(test_case) :
    left_stack = []
    right_stack = []
    data = input() 
    for str in data : 
        if str == '-' : 
            if left_stack : 
                left_stack.pop()
        elif str == '<' : 
            if left_stack :
                right_stack.append(left_stack.pop())
        elif str == '>' :
            if right_stack :
                left_stack.append(right_stack.pop())
        else :
            left_stack.append(str)
            
    left_stack.extend(reversed(right_stack))
    print(''.join(left_stack))
    
```
