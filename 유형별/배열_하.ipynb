# 음계

```python
a = list(map(int, input().split(' ')))
asc = True 
desc = True 

for i in range(1, len(a)) :
    if a[i] > a[i-1] : 
        desc = False
    elif a[i] < a[i-1] : 
        asc = False

if asc :
    print('ascending')
elif desc : 
    print('descending')
else : 
    print('mixed')
```

# 블랙잭

```python
def near(lst, m) : 
    best = 0
    min = max(lst) 
    for i in range(len(lst)) : 
        cal = abs(lst[i]-m)
        if cal < min : 
            best = i 
            min = cal
    return lst[best]

from itertools import combinations
n, m = list(map(int, input().split(' ')))
data = list(map(int, input().split(' ')))

total = ''
for i in range(len(data)) :
    total += str(data[i])

sum_ = []
for comb in combinations(total, 3) : 
    s = 0
    for i in comb :
        s += int(i) 
    sum_.append(s)
answer = near(sum_, m)
print(answer)
```

```python
n, m = list(map(int, input().split(' ')))
data = list(map(int, input().split(' ')))

result = 0 
length = len(data)

for i in range(0, length) : 
    for j in range(i+1, length) : 
        for k in range(j+1, length) : 
            sum_value = data[i] + data[j] + data[k]
            if sum_value <= m : 
                result = max(result, sum_value)
print(result)
```
