```python
N, lst = input(), list(map(int, input().split()))

for i in range(1, len(lst)) : 
    lst[i] *= (i+1)

ans = [lst[0] for i in range(len(lst))]
for i in range(1, len(lst)) : 
    ans[i] = lst[i] - lst[i-1]
    
for i in ans : 
    print(i, end = ' ')
```
