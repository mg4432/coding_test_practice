```python
from functools import cmp_to_key

def compare(a, b) : 
    if a[1] == b[1] : # value  key    , (a, b)  ,    
        return a[0] - b[0]
        
    else : 
        # value  value  , a[1] > b[1]  a[1] - b[1]   (b, a).    
        # b[1] - a[1]      (a, b)
        return b[1] - a[1] 
        

def solution(N, stages):
    answer = []
    #    
    total = len(stages)
    users = [0 for _ in range(N+1)] # N   
    fails = [] #   list 
    
    # stage user   users[0] = stage 1  , ,...
    for s in stages : 
        users[s-1]  += 1  
    
    #   
    for i in range(N) : 
        if users[i] == 0 : #     
            fails.append((i+1, 0)) 
            
        else : 
            fails.append((i+1, users[i] / total)) #  i    i+1 
            #      
            total -= users[i] 
            

    # fails          
    
    for f in sorted(fails, key = cmp_to_key(compare)) : 
        answer.append(f[0])

    return answer


solution(5, [2,1,2,6,2,4,3,3])
```
