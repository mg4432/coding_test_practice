```python
from functools import cmp_to_key

def compare(a, b) : 
    if a[1] == b[1] : # value가 같은 경우 key 순서대로, a의 key가 더 크면 양수 -> 순서 변동
        return a[0] - b[0] 
    else :
        return b[1] - a[1] # b의 value가 크면 양수, (b, a) 로 순서 변동 
    
    
def solution(N, stages) : 
    answer = [] 
    
    # total user 
    total = len(stages) 
    # user수 입력할 list 선언 
    users = [0 for _ in range(N+1)] # 모든 스테이지를 클리어한 경우까지 포함해서 N+1 
    # 실패율 입력할 list 선언 
    fails = [] 
    
    for s in stages :
        # index는 0, 1, 2, ...지만 1, 2, 3으로 입력
        users[s-1] += 1
    
    for i in range(N) : 
        # 실패한 유저 없는 경우 
        if users[i] == 0 : 
            fails.append((i+1, 0)) 
        # 실패한 유저 있는 경우 실패율을 계산할 수 있음 
        else : 
            fails.append((i+1, users[i]/total))
            total -= users[i]
            
    # fails 계산 완료 -> 비교 
    
    for f in sorted(fails, key = cmp_to_key(compare)) :
        answer.append(f[0]) # 정렬된 key 입력 
        
    return answer 
```

```python
from functools import cmp_to_key

def compare(a, b) : 
    if a[1] == b[1] : # value 같으면 key값 작은 튜플이 앞으로 즉, (a, b) 순서 유지, 음수 반환돼야 함 
        return a[0] - b[0]
        
    else : 
        # value 다르면 value큰 값이 앞으로, a[1] > b[1] 이면 a[1] - b[1]은 양수 반환 (b, a). 즉 음수 반환돼야 함
        # b[1] - a[1] 은 음수이므로 순서 바뀌지 않아서 (a, b)
        return b[1] - a[1] 
        

def solution(N, stages):
    answer = []
    # 전체 유저 수 
    total = len(stages)
    users = [0 for _ in range(N+1)] # N스테이지까지 모두 클리어한 경우
    fails = [] # 실패율 입력할 list 
    
    # stage별 user 수  users[0] = stage 1 실패 , ,...
    for s in stages : 
        users[s-1]  += 1  
    
    # 실패율 계산 
    for i in range(N) : 
        if users[i] == 0 : # 실패한 유저 없는 경우 
            fails.append((i+1, 0)) 
            
        else : 
            fails.append((i+1, users[i] / total)) # 인덱스는 i이지만 실제 스테이지 번호는 i+1 
            # 실패율 계산했으면 해당 유저는 제외 
            total -= users[i] 
            

    # fails 계산 완료        
    
    for f in sorted(fails, key = cmp_to_key(compare)) : 
        answer.append(f[0])

    return answer


solution(5, [2,1,2,6,2,4,3,3])
```
