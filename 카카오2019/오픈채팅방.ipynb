```python
def solution(record) : 
    answer = [] 
    idmap = {}
    
    for str in record : 
        strlist = str.split()
        cmd = strlist[0] # enter, leave, change 
        if cmd == 'Enter' or cmd == 'Change' : 
            id = strlist[1]
            name = strlist[2] 
            
            idmap[id] = name 
    # for문을 전부 순회하고 나면 id에 대한 최종 닉네임이 기록된다
    
    # 채팅방에 최종적으로 남을 메시지
    for str in record : 
        strlist = str.split() 
        cmd = strlist[0]
        if cmd == 'Enter' : 
            id = strlist[1] 
            answer.append(idmap[id] + '님이 들어왔습니다.')
        elif cmd == 'Leave' : 
            id = strlist[1] 
            answer.append(idmap[id] + '님이 나갔습니다.')
    return answer
```
